```python
import pandas as pd

# Load CSV file
df = pd.read_csv("fish_feeding_data.csv")  # Change to 'fish_feeding_data.xlsx' if using Excel

# Display first few rows
print(df.head())
```

```python
# Summary statistics
print(df.describe())

# Check correlation between feeding and growth
correlation = df['Feed Quantity (g)'].corr(df['Fish Growth (cm)'])
print(f"ðŸ“Š Correlation between Feed Quantity and Fish Growth: {correlation:.2f}")
```

```python
import matplotlib.pyplot as plt

# Plot Feeding vs. Growth
plt.figure(figsize=(8,5))
plt.plot(df['Day'], df['Fish Growth (cm)'], marker='o', linestyle='-', color='b', label='Growth Rate')

# Labels and title
plt.xlabel('Day')
plt.ylabel('Fish Growth (cm)')
plt.title('Fish Growth vs. Feeding Quantity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data
X = df['Feed Quantity (g)'].values.reshape(-1, 1)
y = df['Fish Growth (cm)'].values.reshape(-1, 1)

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict Growth for New Feeding Quantities
new_feed = np.array([[30], [35], [40]])  # Predict for feed 30g, 35g, 40g
predicted_growth = model.predict(new_feed)

# Display predictions
for feed, growth in zip(new_feed, predicted_growth):
    print(f"ðŸ€ For Feed Quantity {feed[0]}g, Predicted Growth = {growth[0]:.2f} cm")
```

```python
# Define a threshold
growth_threshold = 2.0  # If growth is below 2 cm, alert the farmer

# Find days with low growth
low_growth_days = df[df['Fish Growth (cm)'] < growth_threshold]['Day']

# Alert the farmer
if not low_growth_days.empty:
    print(f"âš ï¸ Alert: Fish growth is low on days: {list(low_growth_days)}. Check feeding strategy!")
```

```python
# Save Data to Excel with Predictions
df['Predicted Growth (cm)'] = model.predict(df[['Feed Quantity (g)']])
df.to_excel("optimized_fish_feeding.xlsx", index=False)

print("âœ… Optimized feeding schedule saved to 'optimized_fish_feeding.xlsx'")
```
